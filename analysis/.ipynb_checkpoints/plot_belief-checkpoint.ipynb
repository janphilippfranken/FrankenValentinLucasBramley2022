```python
import numpy as np
import pylab
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
from matplotlib.lines import Line2D
from scipy.stats import beta
import seaborn as sb
from bayesian_learner import BayesianLearner
```

```python
x = np.linspace(0.01,.99,140)
belief = beta.ppf(x, 2, 2)
BL = BayesianLearner(2,2,7)
bar_probs = BL.discrete_beta()
```

```python
# GLOBAL PLOT PARAMETERS
plot_params = {'legend.fontsize': 'large',
               'axes.labelsize': 'large',
               'axes.titlesize':'20',
               'axes.labelsize':'20',
               'xtick.labelsize':'20',
               'font.family': 'Arial',
               'xtick.color':'grey',
               'ytick.color':'grey',
               'ytick.labelsize':'20',
               'axes.linewidth': '3'}



pylab.rcParams.update(plot_params)


def plot_belief(data,bar_probs):
    fig, axs = plt.subplots(2, 1, figsize=(5,4))
    struct_idx = 0
  
    n_bins = 7
    start_bin = 0
    x = np.linspace(0,1,140)
    bin_alphas = [0.4,0.4,0.4,0.6,0.4,0.4,0.4]
    bin_colors = ['deepskyblue','deepskyblue','deepskyblue','grey','red','red','red']
    bin_step = int(round(140/n_bins))
    
    axs[0].plot(x, beta.pdf(x, 3, 2), 'black', lw=0, alpha=0.9, label='beta pdf')
    
    for i in range(7):
        axs[0].fill_between(x[start_bin:bin_step], alpha=bin_alphas[i], y1=beta.pdf(x[start_bin:bin_step], 3, 2), facecolor=bin_colors[i],  path_effects=[pe.Stroke(linewidth=2, foreground='black'), pe.Normal()])
        start_bin += int(round(140/n_bins))
        bin_step +=  int(round(140/n_bins))
    
    
       
        
    
    axs[1].bar(x=[0],height=bar_probs[0],path_effects=[pe.Stroke(linewidth=5, foreground='black'), pe.Normal()],
           color='white', width=.7)
    axs[1].bar(x=[1],height=bar_probs[1],path_effects=[pe.Stroke(linewidth=5, foreground='black'), pe.Normal()],
           color='white', width=.7)
    axs[1].bar(x=[2],height=bar_probs[2],path_effects=[pe.Stroke(linewidth=5, foreground='black'), pe.Normal()],
           color='white', width=.7)
    axs[1].bar(x=[3],height=bar_probs[3],path_effects=[pe.Stroke(linewidth=5, foreground='black'), pe.Normal()],
           color='white', width=.7)
    axs[1].bar(x=[4],height=bar_probs[4],path_effects=[pe.Stroke(linewidth=5, foreground='black'), pe.Normal()],
           color='white', width=.7)
    axs[1].bar(x=[5],height=bar_probs[5],path_effects=[pe.Stroke(linewidth=5, foreground='black'), pe.Normal()],
           color='white', width=.7)
    axs[1].bar(x=[6],height=bar_probs[6],path_effects=[pe.Stroke(linewidth=5, foreground='black'), pe.Normal()],
           color='white', width=.7)
    
    
    
    
    min_y = min(beta.pdf(x, 2, 2))
    max_y = max(beta.pdf(x, 2, 2))
    # bins = [1/7, 2/7, 3/7, 4,] 
    axs[0].set_xticks([])
    axs[0].set_xticklabels([]) 

    axs[0].axvline(x=0, label=r'$\mu$', color='black', linestyle='--')
    axs[0].set_xlim([0, 1])
    axs[0].set_ylim([-0.2,2.1])
  
    axs[0].set_yticks([0,1,2])
    axs[0].set_yticklabels([0,1,2])
   
    axs[0].axhline(y=0, color='black')
   

    axs[1].spines['right'].set_visible(False)
    axs[1].spines['top'].set_visible(False)
    axs[1].spines['bottom'].set_linewidth(3)
    axs[1].spines['left'].set_linewidth(3)
    axs[0].spines['right'].set_visible(False)
    axs[0].spines['top'].set_visible(False)
    axs[0].spines['bottom'].set_linewidth(3)
    axs[0].spines['left'].set_linewidth(3)
    axs[1].set_ylabel("Probability")  
    axs[0].set_ylabel("Density")  
    axs[0].set_xlabel(r"$Beta(\alpha =3,\beta =2)$")  
    axs[1].set_xticks([0,1,2,3,4,5,6])
    axs[1].set_xticklabels([])
    # axs.axvline(x=.5, label=r'$\mu$', color='black', linestyle='--')
    

    fig.tight_layout()
    # plt.savefig('descriptive_plots/belief.pdf') 
    plt.show()

plot_belief(belief,bar_probs)
```

```python
x = np.linspace(0.01,.99,100)
belief = beta.ppf(x,2,2)
BL = BayesianLearner(2,2,2)
bar_probs = BL.discrete_beta()
```

```python
# GLOBAL PLOT PARAMETERS
plot_params = {'legend.fontsize': 'large',
               'axes.labelsize': 'large',
               'axes.titlesize':'20',
               'axes.labelsize':'20',
               'xtick.labelsize':'20',
               'font.family': 'Arial',
               'xtick.color':'grey',
               'ytick.color':'grey',
               'ytick.labelsize':'20',
               'axes.linewidth': '3'}

pylab.rcParams.update(plot_params)


def plot_belief(a, b):
    fig, axs = plt.subplots(1, 1, figsize=(7,4))
    struct_idx = 0
    n_bins = 2
    start_bin = 0
    x = np.linspace(0.01,.99,100)
    bin_alphas = [0.4,0.4]
    bin_colors = ['deepskyblue','red']
    bin_step = int(round(100/n_bins))
    axs.plot(x, beta.pdf(x, a, b), 'black', lw=0, alpha=0.9, label='beta pdf')
    for i in range(2):
        axs.fill_between(x[start_bin:bin_step], alpha=bin_alphas[i], y1=beta.pdf(x[start_bin:bin_step], a, b), facecolor=bin_colors[i],  path_effects=[pe.Stroke(linewidth=2, foreground='black'), pe.Normal()])
        start_bin += int(round(100/n_bins))
        bin_step +=  int(round(100/n_bins))
    
    
    min_y = min(beta.pdf(x, a, b))
    max_y = max(beta.pdf(x, a, b))
    axs.axvline(x=a/(a+b), label=r'$\mu$', color='black')
    axs.set_xlim([0, 1])
    axs.set_ylim([-0.1,max_y])
    axs.set_yticks([0,1,2])
    axs.set_yticklabels([0,1,2]) 
    axs.axhline(y=0, color='black')
    axs.spines['right'].set_visible(False)
    axs.spines['top'].set_visible(False)
    axs.spines['bottom'].set_linewidth(3)
    axs.spines['left'].set_linewidth(3)
    axs.set_ylabel("Density")  
    axs.set_xlabel(r"$Beta(\alpha =3,\beta =2)$")  
    # axs.set_xticks([0,1])
    # axs.set_xticklabels([])

    fig.tight_layout()
    plt.show()
```

```python
# 1 = heads 
# 0 = tails 
```

```python
# prior belief 
BL = BayesianLearner(prior_a=1, prior_b=1, n_choices=2)
plot_belief(BL.a, BL.b)
```
